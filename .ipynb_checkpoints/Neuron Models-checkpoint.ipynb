## Hodgkin-Huxley Model

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from scipy import integrate
import hh_model
```

Let:
- $V$ be voltage (membrane potential),
- $J_\mathrm{M}$ is the total current across the membrane
- $V_\mathrm{L}$, $V_\mathrm{Na}$, $V_\mathrm{K}$ zero-current potentials for leak, sodium, and potassium
- $g_\mathrm{L}$, $g_\mathrm{Na}$, $g_\mathrm{K}$ conductance for leak, sodium, and potassium channels
- $n^4$ fraction of potassium channels open
- $m^3 h$ fraction of sodium channels open
- $n_\infty$, $m_\infty$, and $h_\infty$ are equilibrium solutions for $n$, $m$, and $h$, respectively.
- $c$, $\tau_n$, $\tau_m$, $\tau_h$ are time constants


\begin{align}
c \frac{dV}{dt} &= J_\mathrm{M} - g_\mathrm{L} \left( V - V_\mathrm{L} \right) - g_\mathrm{Na} m^3 h \left( V - V_\mathrm{NA} \right) - g_\mathrm{K} n^4 \left(V - V_\mathrm{K} \right) \\
\tau_n \frac{dn}{dt} &= n_\infty - n \\
\tau_m \frac{dm}{dt} &= m_\infty - m \\
\tau_h \frac{dh}{dt} &= h_\infty - h
\end{align}

### Voltage-Dependent Gates

Plot $n(V)$, $m(V)$, and $h(V)$, etc.

```python
V = np.linspace(-100,50,200)
plt.figure()
plt.plot(V, [hh_model.nGate(v)[0] for v in V])
plt.plot(V, [hh_model.mGate(v)[0] for v in V])
plt.plot(V, [hh_model.hGate(v)[0] for v in V])
plt.xlabel('Voltage (mV)')
plt.ylabel('Fraction of Gates Open')
plt.legend(('$n_\infty$','$m_\infty$','$h_\infty$'), fontsize='x-large');
```

```python
plt.figure()
plt.plot(V, [hh_model.nGate(v)[1] for v in V])
plt.plot(V, [hh_model.mGate(v)[1] for v in V])
plt.plot(V, [hh_model.hGate(v)[1] for v in V])
plt.xlabel('Voltage (mV)')
plt.legend((r'$\tau_n$',r'$\tau_m$',r'$\tau_h$'), fontsize='x-large');
```

### Demonstrations

```python
# Set the time span for the simulation (in ms)
dt = 0.01 # length of time step, in units of ms
t = np.arange(0.0, 800.0, dt)  # Again, in units of ms

# Set the initial state vector
#            V              m           h            n
z0 = [-53.14669365,   0.18920957,   0.21004222,   0.50796341]

# Define a function that evaluates the injected current at any time t.
def InjectedCurrent(t):
    '''
    I = InjectedCurrent(t)
    where t is time (measured in ms)
    and I is the current (measured in mA)
    '''
    if t>0.0 and t<200.:
        return -0.2
    elif t<400.:
        return 0.
    elif t<600.:
        return 0.1
    else:
        return 0.6

J = [InjectedCurrent(tt) for tt in t]
plt.figure()
plt.plot(t, J), plt.grid('on');
```

```python
# Run the numerical solver, using HHDynamics as the dynamics
# function.  The output is a matrix of states (each row
# corresponding to one time-stamp).

y = integrate.odeint(hh_model.HHDynamics, z0, t, args=(InjectedCurrent,))
```

```python
# Now plot the results

plt.figure(), plt.clf()
plt.plot(t,y[:,0])
plt.ylabel('V')
plt.xlabel('t'); # label the vertical axis
```

```python
plt.figure()
plt.plot(t,y[:,1], t,y[:,2], t,y[:,3]);
plt.ylabel('Fraction of Gates Open')
plt.legend((r'$m$',r'$h$',r'$n$'), fontsize='x-large');
```

## Leaky Integrate-and-Fire Neuron

### Introduction

Similar to HH model, but simpler. The LIF model does not model the spiking behaviour, just the sub-threshold voltage. When the membrane potential reaches the threshold, we record a spike (as a time stamp).

The sub-threshold membrane potential is governed by the DE,
$\def\L{\mathrm{L}}$
$$
c \frac{dV}{dt} = J_\mathrm{in} - g_\L (V -V_\L) \ , \quad V<V_\L
$$
where $c$ is the membrane capacitance, $J_\mathrm{in}$ is the membrane current, $g_\L$ is the leak conductance, and $V_\L$ is the zero-current potential.

Using the change of variables, $v = \frac{V - V_\L}{V_\mathrm{th} - V_\L}$, where $V_\mathrm{th}$ is the voltage threshold for spiking, and Ohm's Law ($V = JR$), our DE can be written
$$
\tau_m \frac{dv}{dt} = v_\mathrm{in}(t) - v
$$
where $\tau_m$ is the membrane time constant, and $v_\mathrm{in}$ is the input voltage (current).

```python
import jonef
%matplotlib inline
```

```python
# Integration parameters
P = 1000
dt = 0.001
T = P*dt
tt = np.linspace(0, T, P, endpoint=True)
```

### Input Current

```python
# Constant input (in the form of a time series)
vin = 0.8*np.ones(shape=[P,1])
```

```python
# Simple sine curve
vin = 1. + 0.4*np.sin(tt*4.*np.pi)
```

```python
plt.figure()
plt.plot(tt,vin, 'y'), plt.plot(tt, np.ones(np.shape(tt)), 'r--')
plt.ylabel('Input Current (J)'); plt.ylim([-0.5,2]);
```

```python
# Feed the current into a LIF neuron
spike_times, v = jonef.Current2Spikes(vin[:,np.newaxis], dt)
```

### Plot Neuron Activity

```python
plt.figure()
plt.plot(tt, v[:,0])
plt.plot(tt, vin, 'y')

jonef.PlotSpikeRaster(spike_times, y_range=[1.5,1.7])
plt.plot(tt, np.ones(np.shape(tt)), 'r--')
plt.ylim([-0.5,2])
plt.xlabel('Time (s)')
plt.ylabel('Voltage (v) or Input Current (vin)');
```

### Spike Rate

What if we hold the input current ($v_\mathrm{in}$) constant and measure the steady-state spike rate?

```python
vin = 2.1
spike_times, v = jonef.Current2Spikes(np.ones(shape=[P,1])*vin, dt)
plt.figure()
jonef.PlotSpikeRaster(spike_times, y_range=[0.5,0.6])
plt.xlabel('Time (s)');plt.ylim([0,1]);
```

```python
print('Input current = ' + str(vin))
n_spikes = len(spike_times[0])
print(str(n_spikes) + ' spikes in ' + str(T) + ' seconds')
print('Spike rate = ' + str(n_spikes/T) + ' Hz')
```

#### Plot of input current vs. spike rate

```python
vin_vals = np.linspace(0., 3., num=50)
spike_rates = []
for vin in vin_vals:
    spike_times, v = jonef.Current2Spikes(np.ones(shape=[P,1])*vin, dt, interp=True)
    n_spikes = len(spike_times[0])
    spike_rates.append(n_spikes)
plt.figure()
plt.plot(vin_vals, spike_rates, '.')
plt.ylabel('Firing Rate (Hz)')
plt.xlabel('Input ($v_\mathrm{in}$)');
```

### LIF Tuning Curve

```python
def G(vin):
    tau_ref = 0.002
    tau_m = 0.02
    if vin>1.:
        return 1./(tau_ref - tau_m*np.log(1.-1./vin))
    else:
        return 0.
```

```python
a = [G(v) for v in vin_vals]
plt.figure()
plt.plot(vin_vals, a, 'r')
plt.plot(vin_vals, spike_rates, '.', color='C0')
plt.xlabel('Input ($v_\mathrm{in}$)'), plt.ylabel('Firing Rate (Hz)');
```

## Tuning Curves / Activation Functions

### Logistic

Given input $z$, the activity of the sigmoid neuron is $\sigma (z)$, where $\sigma$ is

$$
\sigma (z) = \frac{1}{1 + \exp (-z)}
$$

```python
def logistic(z):
    return 1./(1+np.exp(-z))
```

```python
z = np.linspace(-6., 6, num=201)
h1 = logistic(z)
plt.figure()
plt.plot(z,h1)
plt.grid('on');
```

### Arctan

$$
\sigma (z) = \arctan (z)
$$

```python
def arctan(z):
    return np.arctan(z)
```

```python
plt.figure()
plt.plot(z, arctan(z)); plt.grid('on');
```

### Hyperbolic Tangent

$$
\sigma (z) = \tanh (z)
$$

```python
def tanh(z):
    return np.tanh(z)
```

```python
plt.figure()
plt.plot(z, tanh(z)); plt.grid('on');
```

### Rectified Linear Unit (ReLU)

This is a linear unit that is capped below at zero.
$$
\sigma (z) = \max (0, z)
$$

```python
def ReLU(z):
    return np.maximum(0., z)
```

```python
plt.figure()
plt.plot(z, ReLU(z)); plt.grid('on');
```
